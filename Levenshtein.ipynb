```python
def levenshtein_recursif(a, b):
    global count
    count +=  1
    i, j = len(a), len(b)
    if i == 0 :
        return j
    elif j == 0 :
        return i
    elif a[i - 1] == b[j - 1]:
        return levenshtein_recursif(a[:i-1], b[:j-1])
    else :
        return min(levenshtein_recursif(a[:-1], b[:]), levenshtein_recursif(a[:], b[:-1]), levenshtein_recursif(a[:-1], b[:-1])) + 1

a = ["hello", "hello", "hey",   ""        , "magic"]
b = ["hello", "hella", "hello", "hello"   , "citric"]
expected = [0, 1, 3, 5, 4]

count = 0
for i in range(len(expected)) :
    print(f"[{a[i]}] - [{b[i]}] : {expected[i]}")
    assert levenshtein_recursif(a[i], b[i]) == expected[i]
    print(f"[{b[i]}] - [{a[i]}] : {expected[i]}")
    assert levenshtein_recursif(b[i], a[i]) == expected[i]

print("nb appels", count)
```

```python
def levenshtein_memoise(a, b, dico = None):
    if dico == None:
        dico = {}
    global count
    count += 1
    i, j = len(a), len(b)
    
    # regarde si deja dans le dico
    if (i,j) in dico.keys():
        # print("Utilise memoisation")
        return dico[(i,j)]
    
    if i == 0 :
        dico[(i,j)] = j
    elif j == 0 :
        dico[(i,j)] = i
    elif a[i - 1] == b[j - 1]:
        dico[(i,j)] = levenshtein_memoise(a[:-1], b[:-1], dico)
    else :
        dico[(i,j)] = min(levenshtein_memoise(a[:-1], b[:], dico), 
                          levenshtein_memoise(a[:], b[:-1], dico), 
                          levenshtein_memoise(a[:-1], b[:-1], dico)) + 1
        
    return dico[(i,j)]

count = 0
for i in range(len(expected)) :
    print(f"[{a[i]}] - [{b[i]}] : {expected[i]}")
    assert levenshtein_memoise(a[i], b[i]) == expected[i]
    print(f"[{b[i]}] - [{a[i]}] : {expected[i]}")
    assert levenshtein_memoise(b[i], a[i]) == expected[i]
    
print("nb appels", count)
```

```python
def levenshtein_bottomup(a, b):
    i, j = len(a), len(b)
    d = [[0 for jj in range(j + 1)] for ii in range(i+1)] # i ligne, j colonne

    for ii in range(i+1):
        d[ii][0] = ii   # correspond à j = 0
    for jj in range(j+1):
        d[0][jj] = jj  # correspond à i = 0
    
    for ii in range(1, i+1):  
        for jj in range(1, j+1):
            if a[ii-1] == b[jj-1]:
                d[ii][jj] = d[ii-1][jj-1]
            else:
                d[ii][jj] = min(d[ii-1][jj],  
                         d[ii][jj-1],                   
                         d[ii-1][jj-1]) +1

    return d[i][j]    
  
for i in range(len(expected)) :
    print(f"[{a[i]}] - [{b[i]}] : {expected[i]}")
    assert levenshtein_bottomup(a[i], b[i]) == expected[i]
    print(f"[{b[i]}] - [{a[i]}] : {expected[i]}")
    assert levenshtein_bottomup(b[i], a[i]) == expected[i]
```

```python
def levenshtein_bottomup_less_space(a, b):
    i, j = len(a), len(b)
    
    d_iprev = [jj for jj in range(j+1)]
    d_i = d_iprev[:]
    
    for ii in range(1, i+1):  
        d_i[0] = ii
        for jj in range(1, j+1):
            if a[ii-1] == b[jj-1]:
                d_i[jj] = d_iprev[jj-1]
            else:
                d_i[jj] = min(d_iprev[jj],  
                         d_i[jj-1],                   
                         d_iprev[jj-1]) +1
                
        d_iprev = d_i[:]
    return d_i[j]    
  

for i in range(len(expected)) :
    print(f"[{a[i]}] - [{b[i]}] : {expected[i]}")
    assert levenshtein_bottomup_less_space(a[i], b[i]) == expected[i]
    print(f"[{b[i]}] - [{a[i]}] : {expected[i]}")
    assert levenshtein_bottomup_less_space(b[i], a[i]) == expected[i]
```
